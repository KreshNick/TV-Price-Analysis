```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
m65inch = pd.read_csv("m75inch.csv")
```

```python
m65inch
```

```python
ax = m65inch[["Cmimi minimal", "Cmimi maksimal", "Cmimi mesatar"]].plot(kind='bar',
              figsize=(15,7), color=['red', 'green', 'blue'], fontsize=13);
ax.set_alpha(0.8)
ax.set_title("TV 75 inch",
fontsize=18)
ax.set_ylabel("Cmimi", fontsize=18);
ax.set_yticks([100, 300, 500, 700, 900, 1500, 2000, 2500, 3000, 4000])
ax.set_xticklabels(["Aztech", "Gama", 
                    "Gjirafa", 
                    "Interplast", "Neptun"], rotation=0, fontsize=11)

for i in ax.patches:
    ax.text(i.get_x()+.04, i.get_height()+5, \
            str(round((i.get_height()), 2)), fontsize=11, color='black',
                rotation=60)
```
