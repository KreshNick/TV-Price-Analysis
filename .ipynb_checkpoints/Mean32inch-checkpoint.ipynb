```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
m50inch = pd.read_csv("m50inch.csv")
```

```python
m50inch
```

```python
ax = m50inch[["Cmimi minimal", "Cmimi maksimal", "Cmimi mesatar"]].plot(kind='bar',
              figsize=(15,7), color=['red', 'green', 'blue'], fontsize=13);
ax.set_alpha(0.8)
ax.set_title("TV 48-50 inch",
fontsize=18)
ax.set_ylabel("Cmimi", fontsize=18);
ax.set_yticks([100, 200, 300, 400, 500, 600, 700, 800, 900, 1000])
ax.set_xticklabels(["Aztech", "Buton", "Gama", 
                    "Gjirafa", "Gorenje", 
                    "Interplast", "Neptun"], rotation=0, fontsize=11)

for i in ax.patches:
    ax.text(i.get_x()+.04, i.get_height()+5, \
            str(round((i.get_height()), 2)), fontsize=11, color='black',
                rotation=60)
```
